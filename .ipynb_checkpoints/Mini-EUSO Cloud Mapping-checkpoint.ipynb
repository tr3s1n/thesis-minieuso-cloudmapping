### Function for stripping bits used for reading MODIS cloud mask values

```python
def bits_stripping(bit_start, bit_count, value):
    bitmask=pow(2, bit_start+bit_count)-1
    return np.right_shift(np.bitwise_and(value,bitmask),bit_start)
```

###  MODIS data test cropping to bounding box with rioxarray
- first we need to define our bounding box used for cropping with shapely.geometry.box
- in this case we are using the last bounding box in the list, so its the FoV in the last timestamp of mini-EUSO data

```python
def plot_modis_cloud_mask(modis_cm_clip):
    binary_mask = []
    for pixel in modis_cm_clip.values[0].reshape(-1):
        if int(pixel) == 0:
            binary_mask.append(0)
        else:
            if bits_stripping(1,2, pixel) <= 1:
                binary_mask.append(1)
            else:
                binary_mask.append(0)

    binary_mask_arr = np.asarray(binary_mask)
    binary_mask_arr_reshaped = binary_mask_arr.reshape((len(modis_cm_clip.y.values), len(modis_cm_clip.x.values)))
    
    plt.figure(figsize=(8,8))
    im = plt.imshow(binary_mask_arr_reshaped, extent = [modis_cm_clip.x.values.min(), modis_cm_clip.x.values.max(), modis_cm_clip.y.values.min(), modis_cm_clip.y.values.max()], interpolation='none')
    values = np.array([0,1])
    colors = [(0.267004, 0.004874, 0.329415, 1.0), (0.993248, 0.906157, 0.143936, 1.0)]
    patches = [ mpatches.Patch(color=colors[0], label="Clear".format(l=values[0])),
                mpatches.Patch(color=colors[1], label="Cloudy".format(l=values[1])) ]
    plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0. )
    plt.xlabel("Longitude [deg]", fontweight ='bold')
    plt.ylabel("Latitude [deg]", fontweight ='bold')
    plt.title('MODIS L2 Cloud Mask 1km', fontweight ="bold")
    plt.show()
```

### Test mapping function with all the results for each pixel stored in separate lists

```python
start_time = datetime.now()
test_res_mask = []
test_minieuso_pixel_edges_list = []
test_minieuso_pixel_center_list = []
test_modis_cloud_mask_clips_list = []
test_modis_pixels_center_list = []
test_modis_pixels_values_list = []
test_all_modis_pixels_center_list = []

for pixel_edges_arr in test_pixels_edges.reshape((-1, 5, 3)):
    pixel_lons = []
    pixel_lats = []
    curr_pixel_edges_latlon = []
    
    for edge in pixel_edges_arr[1:]:
        pixel_edge_geographic_position = geographic_position_of_distance(
            skyfield.toposlib.wgs84, 
            skyfield.units.Distance(km=edge))
        pixel_lons.append(pixel_edge_geographic_position.longitude.degrees)
        pixel_lats.append(pixel_edge_geographic_position.latitude.degrees)
        curr_pixel_edges_latlon.append((pixel_edge_geographic_position.latitude.degrees, pixel_edge_geographic_position.longitude.degrees))
    
    pixel_bbox = [box(min(pixel_lons), min(pixel_lats), max(pixel_lons), max(pixel_lats))]
    pixel_center_latlon = (geographic_position_of_distance(skyfield.toposlib.wgs84, skyfield.units.Distance(km=pixel_edges_arr[0])).latitude.degrees, geographic_position_of_distance(skyfield.toposlib.wgs84, skyfield.units.Distance(km=pixel_edges_arr[0])).longitude.degrees)
    
    modis_cloud_mask_clip = rxr.open_rasterio(test_modis_tif_cm_file, masked=True).rio.clip(pixel_bbox,
                                                                                   all_touched=True,
                                                                                   from_disk=True)
    
    
    test_minieuso_pixel_center_list.append(pixel_center_latlon)
    test_minieuso_pixel_edges_list.append(curr_pixel_edges_latlon)
    test_modis_cloud_mask_clips_list.append(modis_cloud_mask_clip)
    
    all_curr_modis_pixels_center_list = []
    curr_modis_pixels_center_list = []
    curr_modis_pixels_bitsvalues_list = []
    pixel_corr_modis_pixels_values_list = []
    
    for iy, ix  in np.ndindex(modis_cloud_mask_clip.values[0].shape):
        current_modis_pixel_center_latlon = (modis_cloud_mask_clip[0].y[iy].item(), modis_cloud_mask_clip[0].x[ix].item())
        all_curr_modis_pixels_center_list.append(current_modis_pixel_center_latlon)
        if geodesic(current_modis_pixel_center_latlon, pixel_center_latlon).kilometers < 2.8:
            if bits_stripping(1,2, modis_cloud_mask_clip[0, iy, ix].item()) <= 1:
                pixel_corr_modis_pixels_values_list.append(1)
            else:
                pixel_corr_modis_pixels_values_list.append(0)
            curr_modis_pixels_bitsvalues_list.append(bits_stripping(1,2, modis_cloud_mask_clip[0, iy, ix].item()))
            curr_modis_pixels_center_list.append(current_modis_pixel_center_latlon)
    
    test_all_modis_pixels_center_list.append(all_curr_modis_pixels_center_list)
    test_modis_pixels_center_list.append(curr_modis_pixels_center_list)
    test_modis_pixels_values_list.append(curr_modis_pixels_bitsvalues_list)
    
    if (sum(pixel_corr_modis_pixels_values_list)/len(pixel_corr_modis_pixels_values_list)) >= 0.5:
        test_res_mask.append(1)
    else:
        test_res_mask.append(0)

print('Time elapsed (hh:mm:ss.ms): {}'.format(datetime.now() - start_time))
```

- function for plotting each minieuso pixel data from the above function

```python
def plot_modis_cloud_mask_and_fov_pixel(cm_clips, pixels_centers_list, pixels_edges, pixel_num):
    
    # binary mask computing
    binary_mask = []
    for pixel in cm_clips[pixel_num].values[0].reshape(-1):
        if int(pixel) == 0:
            binary_mask.append(0)
        else:
            if bits_stripping(1,2, pixel) <= 1:
                binary_mask.append(1)
            else:
                binary_mask.append(0)

    binary_mask_arr = np.asarray(binary_mask)
    binary_mask_arr_reshaped = binary_mask_arr.reshape((len(cm_clips[pixel_num].y.values), len(cm_clips[pixel_num].x.values)))
    
    # pixel outline computing
    pixels_edges_reshaped = pixels_edges.reshape(-1,5,3)
    pixel_edges_list = [pixels_edges_reshaped[pixel_num][3], pixels_edges_reshaped[pixel_num][4], pixels_edges_reshaped[pixel_num][1], pixels_edges_reshaped[pixel_num][2]]
    pixel_edges_lats = []
    pixel_edges_lons = []

    for edge in pixel_edges_list:
        edge_geographic_position = geographic_position_of_distance(
            skyfield.toposlib.wgs84, 
            skyfield.units.Distance(km=edge))
        pixel_edges_lons.append(edge_geographic_position.longitude.degrees)
        pixel_edges_lats.append(edge_geographic_position.latitude.degrees)

    # We need to append the first edge coords to the list again
    edge_geographic_position = geographic_position_of_distance(
        skyfield.toposlib.wgs84, 
        skyfield.units.Distance(km=pixel_edges_list[0]))
    pixel_edges_lons.append(edge_geographic_position.longitude.degrees)
    pixel_edges_lats.append(edge_geographic_position.latitude.degrees)

    pixel_polygon_geom = Polygon(zip(pixel_edges_lons, pixel_edges_lats))
    crs = {'init': 'epsg:4326'}
    pixel_polygon = gpd.GeoDataFrame(index=[0], crs=crs, geometry=[pixel_polygon_geom])
    pixel_polygon_coords = list(pixel_polygon.geometry[0].exterior.coords)
    pixel_x = [i for i,j in pixel_polygon_coords]
    pixel_y = [j for i,j in pixel_polygon_coords]
    
    # pixels centers scatter computing\
    modis_lats = []
    modis_lons = []
    for pixel in pixels_centers_list[pixel_num]:
        modis_lats.append(pixel[0])
        modis_lons.append(pixel[1])
    
    # plotting
    plt.figure(figsize=(10,10))
    
    im = plt.imshow(binary_mask_arr_reshaped, extent = [cm_clips[pixel_num].x.values.min(), cm_clips[pixel_num].x.values.max(), cm_clips[pixel_num].y.values.min(), cm_clips[pixel_num].y.values.max()], interpolation='none')
    plt.plot(pixel_x, pixel_y, 'r-', linewidth=3)
    plt.scatter(modis_lons, modis_lats, s=15, c='b', marker="o")
    
    values = np.array([0,1])
    colors = [(0.267004, 0.004874, 0.329415, 1.0), (0.993248, 0.906157, 0.143936, 1.0)]
    patches = [ mpatches.Patch(color=colors[0], label='Clear'),
                mpatches.Patch(color=colors[1], label='Cloudy'),
                mpatches.Patch(color='r', label='FoV Pixel outline'),
                mpatches.Patch(color='b', label='Assigned MODIS pixels') ]
    plt.legend(handles=patches, bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0. )
    plt.xlabel('Longitude [deg]', fontweight ='bold')
    plt.ylabel('Latitude [deg]', fontweight ='bold')
    plt.title('MODIS L2 Cloud Mask 1km clip with Mini-EUSO pixel outline', fontweight ="bold")
    plt.show()
```

```python
def plot_corresponding_test_pixels(pixel_num):
    print(f'Plotting data for test pixel with index {pixel_num}: ')
    
    modis_lats = []
    modis_lons = []
    pixel_lats = []
    pixel_lons = []
    all_modis_lats = []
    all_modis_lons = []
    
    for pixel in test_modis_pixels_center_list[pixel_num]:
        modis_lats.append(pixel[0])
        modis_lons.append(pixel[1])

    for pixel in test_minieuso_pixel_edges_list[pixel_num]:
        pixel_lats.append(pixel[0])
        pixel_lons.append(pixel[1])

    for pixel in test_all_modis_pixels_center_list[pixel_num]:
        all_modis_lats.append(pixel[0])
        all_modis_lons.append(pixel[1])
        
    fig = plt.figure(figsize=(8,8))
    ax1 = fig.add_subplot(111)

    ax1.scatter(pixel_lons, pixel_lats, s=25, c='b', marker="s", label='minieuso pixel edges')
    ax1.scatter(modis_lons, modis_lats, s=15, c='r', marker="o", label='assigned modis pixels')
    ax1.scatter(all_modis_lons, all_modis_lats, s=3, c='k', marker="o", label='all modis pixels')
    ax1.scatter(test_minieuso_pixel_center_list[pixel_num][1], test_minieuso_pixel_center_list[pixel_num][0], s=25, c='g', marker='s', label='minieuso pixel center')
    ax1.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0. )
    plt.xlabel("Longitude [deg]", fontweight ='bold')
    plt.ylabel("Latitude [deg]", fontweight ='bold')
    plt.title('Mapping of MODIS pixels to Mini-EUSO pixel', fontweight ="bold")
    plt.show()

    print('\n')
    plot_modis_cloud_mask_and_fov_pixel(test_modis_cloud_mask_clips_list, test_modis_pixels_center_list, test_pixels_edges, pixel_num)
    
    average_value = sum([1 if i <= 1 else 0 for i in test_modis_pixels_values_list[pixel_num]])/len([1 if i <= 1 else 0 for i in test_modis_pixels_values_list[pixel_num]])
    print('Average value of modis corresponding MODIS pixels: {}'.format(average_value))
    if average_value >= 0.5:
        print('Pixel was classified as cloudy.')
    else:
        print('Pixel was classified as clear.')
```

### Final mapping function
- takes pixels edges array and modis tif as an input
- output is binary cloud mask of Mini-EUSO pixels (list with 48x48 = 2304 values)

```python
def get_cloudmask_of_minieuso_fov(minieuso_pixels_edges, modis_tif):
    res_mask = []
    
    # For each pixel edges array of pixel of 48x48 Mini-EUSO FoV 
    for pixel_edges_arr in minieuso_pixels_edges.reshape((-1, 5, 3)):
        pixel_lons = []
        pixel_lats = []
        
        # Calculate bounding box of the pixel according to its edges ([0] is center so we can omit it)
        for edge in pixel_edges_arr[1:]:
            pixel_edge_geographic_position = geographic_position_of_distance(
                skyfield.toposlib.wgs84, 
                skyfield.units.Distance(km=edge))
            pixel_lons.append(pixel_edge_geographic_position.longitude.degrees)
            pixel_lats.append(pixel_edge_geographic_position.latitude.degrees)

        # Get bounding box (minLon, minLat, maxLon, maxLat) of given pixel
        pixel_bbox = [box(min(pixel_lons), min(pixel_lats), max(pixel_lons), max(pixel_lats))]
        
        # Get LatLon coordinates of the pixel center         
        pixel_center_latlon = (geographic_position_of_distance(skyfield.toposlib.wgs84, skyfield.units.Distance(km=pixel_edges_arr[0])).latitude.degrees,
                               geographic_position_of_distance(skyfield.toposlib.wgs84, skyfield.units.Distance(km=pixel_edges_arr[0])).longitude.degrees)
        
        # Crop MODIS Cloud Mask data to pixel bounding box with rioxarray
        modis_cloud_mask_clip = rxr.open_rasterio(modis_tif, masked=True).rio.clip(pixel_bbox,
                                                                                   all_touched=True,
                                                                                   from_disk=True)


        pixel_corr_modis_pixels_values_list = []
        
        # Iterate through Modis clip pixels
        for iy, ix  in np.ndindex(modis_cloud_mask_clip.values[0].shape):
            # Get center coordinates of each pixel
            current_modis_pixel_center_latlon = (modis_cloud_mask_clip[0].y[iy].item(), modis_cloud_mask_clip[0].x[ix].item())
            # Calculate geodesic distance of the modis pixel to the center of the minieuso pixel
            # If the distance is less than 2.8km, we count this pixel as a part of minieuso pixel
            if geodesic(current_modis_pixel_center_latlon, pixel_center_latlon).kilometers < 2.8:
                # Read 1 and 2 bits and read the value according to docs: 0,1 - Cloudy, 2,3 - Clear
                if bits_stripping(1,2, modis_cloud_mask_clip[0, iy, ix].item()) <= 1:
                    pixel_corr_modis_pixels_values_list.append(1)
                else:
                    pixel_corr_modis_pixels_values_list.append(0)
        
        # if most of the MODIS pixels assigned to Mini-EUSO pixel are cloudy, we flag this pixel as cloudy
        if (sum(pixel_corr_modis_pixels_values_list)/len(pixel_corr_modis_pixels_values_list)) >= 0.5:
            res_mask.append(1)
        else:
            res_mask.append(0)
    
    # Return the result binary mask (list with 2304 (48x48) values)
    return res_mask
```

- function for plotting final cloud mask obtained by the function above

```python
def plot_minieuso_cloud_mask(mask):
    mask_arr = np.asarray(mask)
    mask_arr_reshaped = mask_arr.reshape((48,48))
   
    plt.figure(figsize=(8,8))
    im = plt.imshow(mask_arr_reshaped, interpolation='none')
    values = np.unique(mask_arr.ravel())
    colors = [ im.cmap(im.norm(value)) for value in values ]
    patches = [ mpatches.Patch(color=colors[0], label="Clear".format(l=values[0])),
                mpatches.Patch(color=colors[1], label="Cloudy".format(l=values[1])) ]
    plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0. )
    plt.title('Mini-EUSO Cloud Mask mapped from MODIS data', fontweight ="bold")
    plt.show()
```

```python
def plot_minieuso_photon_count_pixels(data_index):
    minieuso_data_ts = test_eventD3_ts_arr[data_index]
    minieuso_photon_data = test_eventD3_photon_count_arr[:,0,0,:,:].reshape(3200, -1)[data_index]
    minieuso_pixels_edges_arr = get_pixels_edges_on_surface_arr(minieuso_data_ts)
    minval = np.min(minieuso_photon_data[np.nonzero(minieuso_photon_data)])
    maxval = np.max(minieuso_photon_data[np.nonzero(minieuso_photon_data)])
    minieuso_data_dt = datetime.utcfromtimestamp(minieuso_data_ts)

    norm = mpl.colors.Normalize(vmin=minval, vmax=maxval, clip=True)
    cm_mapper = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.viridis)

    fig, ax = plt.subplots(figsize=(8,8))

    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right',size='5%',pad=0.05)

    i = 0

    for points_arr in minieuso_pixels_edges_arr.reshape((-1, 5, 3)):
        geographic_position = geographic_position_of_distance(
            skyfield.toposlib.wgs84, 
            skyfield.units.Distance(km=points_arr[0])
        )
        lon = geographic_position.longitude.degrees
        lat = geographic_position.latitude.degrees
        c = cm_mapper.to_rgba(minieuso_photon_data[i])


        x = []
        y = []

        for point in points_arr[1:]:
            geographic_position = geographic_position_of_distance(skyfield.toposlib.wgs84, skyfield.units.Distance(km=point))
            x.append(geographic_position.longitude.degrees)
            y.append(geographic_position.latitude.degrees)

        ax.fill(x, y, color=c)
        i+=1

    fig.colorbar(cm_mapper, cax=cax)

    ax.set_xlabel('Longitude [deg]', fontweight ='bold')
    ax.set_ylabel('Latitude [deg]', fontweight ='bold')
    ax.set_title(f'Test Mini-EUSO photon_count_data\n index: {data_index}\n {minieuso_data_dt}', fontweight ='bold')

    plt.show()
```

```python
def plot_minieuso_cloud_mask_data_pixels(mask, data_index):

    minieuso_data_ts = test_eventD3_ts_arr[data_index]
    minieuso_pixels_edges_arr = get_pixels_edges_on_surface_arr(minieuso_data_ts)
    
    fov_polygon = get_minieuso_polygon_from_pixels_edges_on_surface_arr(minieuso_pixels_edges_arr)
    fov_modis_time_polygon_clip = rxr.open_rasterio(test_modis_tif_time_file).rio.clip(fov_polygon.geometry,
                                                                                       all_touched=True,
                                                                                       from_disk=True)
    timestamps_unique_list = sorted(np.unique(fov_modis_time_polygon_clip.values).tolist())
    
    offset_dt = datetime(1993,1,1)
    offset_utc = offset_dt.replace(tzinfo=timezone.utc)
    offset_utc_ts = offset_utc.timestamp()

    utc_dt_start = datetime.utcfromtimestamp(offset_utc_ts + int(timestamps_unique_list[1]))
    utc_dt_end = datetime.utcfromtimestamp(offset_utc_ts + int(timestamps_unique_list[-1]))
    
    fig, ax = plt.subplots(figsize=(8,8))
    i=0

    for points_arr in minieuso_pixels_edges_arr.reshape((-1, 5, 3)):
        geographic_position = geographic_position_of_distance(
            skyfield.toposlib.wgs84, 
            skyfield.units.Distance(km=points_arr[0])
        )
        lon = geographic_position.longitude.degrees
        lat = geographic_position.latitude.degrees
        values = np.array([0,1])
        colors = [(0.267004, 0.004874, 0.329415, 1.0), (0.993248, 0.906157, 0.143936, 1.0)]
        patches = [ mpatches.Patch(color=colors[0], label="Clear".format(l=values[0])),
                    mpatches.Patch(color=colors[1], label="Cloudy".format(l=values[1])) ]
        plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0. )
        if mask[i] == 0:
            c = colors[0]
        else:
            c = colors[1]

        x = []
        y = []

        for point in points_arr[1:]:
            geographic_position = geographic_position_of_distance(skyfield.toposlib.wgs84, skyfield.units.Distance(km=point))
            x.append(geographic_position.longitude.degrees)
            y.append(geographic_position.latitude.degrees)

        ax.fill(x, y, color=c)
        i+=1

    ax.set_facecolor('black')
    ax.set_xlabel('Longitude [deg]', fontweight ='bold')
    ax.set_ylabel('Latitude [deg]', fontweight ='bold')
    plt.title(f'Test Mini-EUSO pixels with mapped MODIS Cloud Mask data\n index: {data_index}\n {utc_dt_start} - {utc_dt_end}', fontweight ='bold')
    plt.show()
```
